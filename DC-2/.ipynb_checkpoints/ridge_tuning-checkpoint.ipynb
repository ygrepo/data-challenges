## Read In The Data <a name="Read"></a>

```python
#col_names =['isbuyer','buy_freq','visit_freq','buy_interval','sv_interval','expected_time_buy','expected_time_visit','last_buy','last_visitmultiple_buy','multiple_visit','uniq_urls','num_checkins','y_buy']
df = pd.read_csv('DScasestudy_1.txt','\t')
print('Number of rows:',len(df))
df.head()
```

```python
# Split into training and test sets
X = df.drop(columns = ['response'])
y = df['response']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print('#X={}, #y={}'.format(X.shape,y.shape))
print('#X_train={}, #y_train={}, #X_test={}, #y_test={}'.format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
```

```python
feature_names = list(X.columns.values)
```

```python
%%time
model = RidgeClassifier()
#model = LogisticRegression(class_weight='balanced')
tuning_parameters = {
    'alpha':np.logspace(-5,5,11),
    'class_weight':['balanced',None]
}

rc_search = RandomizedSearchCV(model, tuning_parameters, cv = 5, n_iter= 5, return_train_score=False, n_jobs=4,
                              random_state = 20)
rc_search.fit(X, y)
print('Best parameters:', rc_search.best_params_)
rc = rc_search.best_estimator_
```

```python
rc.fit(X_train, y_train)
y_pred = rc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
report = classifiers_metric_report([rc], ['Ridge Classifier'], X_test, y_test)
report.round(3)
```

```python
plot_confusion_matrix_with_labels(X_test, y_test, rc, rc.classes_)
```

```python
clf = RidgeClassifier(random_state=444)
clf.get_params().keys()
```

```python
pipe = Pipeline([
    ('oversample', SMOTE(random_state=444)),
    ('clf', RidgeClassifier(random_state=444))
    ])

scorers = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score)
}

refit_score='recall_score'

param_grid = {'clf__alpha':np.logspace(-5,5,11)}

skf = StratifiedKFold(n_splits=10)
grid_search = GridSearchCV(pipe, param_grid, scoring=scorers, refit=refit_score,
                           cv=skf, return_train_score=True, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
y_pred = grid_search.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
report = classifiers_metric_report([grid_search], ['Optimized Ridge Classifier'], X_test, y_test)
report.round(3)
```

```python
plot_confusion_matrix_with_labels(X_test, y_test, grid_search, grid_search.classes_)
```
